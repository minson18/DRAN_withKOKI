```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Load the data from the .npy file
returns = np.load('./returns_1e8.npy')

# Define the sample sizes
sample_sizes = [2**n for n in range(1, 27)]

# Function to compute the average of the average of sampled returns
def compute_avg_of_avg(returns, sample_size):
    averages = []
    for _ in range(5):  # Repeat 5 times
        sample = random.sample(list(returns), sample_size)
        averages.append(np.mean(sample))
    return np.mean(averages)

# Compute averages for different sample sizes
avg_averages = [compute_avg_of_avg(returns, size) for size in sample_sizes]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, avg_averages, marker='o')
plt.xscale('log')
plt.xlabel('Sample Size')
plt.ylabel('Mean Return')
plt.title('Average of Averages of Sampled Returns')
plt.grid(True)
plt.show()
```
